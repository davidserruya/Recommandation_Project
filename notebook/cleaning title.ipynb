```python
df.drop('Unnamed: 0', axis=1, inplace=True, errors='ignore')
```

```python
# suppression parenthÃ¨ses
df['Title'] = df['Title'].str.replace(r'\([^)]*\)', '', regex=True)
```

```python
df
```

```python
def move_last_word_to_beginning(title):
    parts = title.split(', ')
    if len(parts) > 1:
        last_word = parts[-1]
        new_title = last_word + ' ' + ', '.join(parts[:-1])
        return new_title
    else:
        return title


df['Title'] = df['Title'].apply(move_last_word_to_beginning)
```
